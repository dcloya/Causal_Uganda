```r
library(readxl)
library(dplyr)
library(magrittr)
```

```r
columns <- c(X__1 = "Covariate", X__2 = "Control: N", X__3 = "Control: Mean", 
             X__5 = "Hard Skills: N", X__6 = "Hard Skills: Mean",
             X__8 = "Soft Skills: N", X__9 = "Soft Skills: Mean",
             X__11 = "p-value: Hard vs Control", X__12 = "p-value: Soft vs Control", X__13 = "p-value: Hard vs Soft")
```

```r
balance_table1 <- read_excel("SEED_balance_table.xlsx", sheet=1, range = "A5:M31", col_names=FALSE) %>% 

                    # Delete the rows which are all NA's except the variable name (due to blank cells in excel)
                    filter(rowSums(is.na(.)) !=ncol(.)-1) %>%
                    # ncol(.)-1 instead of ncol(.) because the "variable name" wasn't blank for the blank rows
                    # in the spreadsheet corresponding to groupings of variables

                    # Delete the columns which are all NA's (due to blank cells in excel)
                    select(which(colSums(is.na(.)) != nrow(.)) )
                    # which() is used following advice here: https://stackoverflow.com/questions/26875199/dplyr-select-using-logical
                    # select() only works with numbers, not logicals (unlike filter, for whatever reason)

colnames(balance_table1) <- columns
```

```r
balance_table2 <- read_excel("SEED_balance_table.xlsx", sheet=2, range = "A5:M33", col_names=FALSE) %>% 

                    # Delete the rows which are all NA's except the variable name (due to blank cells in excel)
                    filter(rowSums(is.na(.)) !=ncol(.)-1) %>%
                    # ncol(.)-1 instead of ncol(.) because the "variable name" wasn't blank for the blank rows
                    # in the spreadsheet corresponding to groupings of variables

                    # Delete the columns which are all NA's (due to blank cells in excel)
                    select(which(colSums(is.na(.)) != nrow(.)) )
                    # which() is used following advice here: https://stackoverflow.com/questions/26875199/dplyr-select-using-logical
                    # select() only works with numbers, not logicals (unlike filter, for whatever reason)

colnames(balance_table2) <- columns
```
